```python
from flask import Flask, request, jsonify
from ipynb.fs.full.pasarEstoaVSenCasaipynb import get_recommendations
```

```python
app = Flask(__name__)

@app.route('http://localhost:3000/userData/recommendations', methods=['POST'])
def recommendations():
    data = request.get_json()

    if 'name' not in data:
        return jsonify({'error': 'Nombre no proporcionado'}), 400

    name_to_search = data['name']
    recommendations_dict = get_recommendations(name_to_search)

    return jsonify(recommendations_dict)

if __name__ == '__main__':
    app.run(debug=True)
```
